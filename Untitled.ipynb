```python
import pandas as pd


data_path = 'C:/Users/Dell/Drugveriseti/Drug.csv'


drugset = pd.read_csv(data_path)


feature_set = ['Age', 'Gender', 'BP', 'Cholesterol', 'Na', 'K', 'Drug']
features = drugset[feature_set] 
target = drugset['Drug']  


print("Features:\n", features.head())
print("\nTarget:\n", target.head())
```

```python
from sklearn.model_selection import train_test_split
feature_train,feature_test, target_train, target_test = train_test_split(features , target , test_size=0.3, random_state=1)
print("feature_train boyutu:", len(feature_train))
print("feature_test boyutu:", len(feature_test))
print("target_train boyutu:", len(target_train))
print("target_test boyutu:", len(target_test))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

feature = drugset.drop('Drug', axis=1)
target = drugset['Drug']


numerical_features = ['Age', 'Na', 'K']
categorical_features = ['Gender', 'BP', 'Cholesterol']


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  
)


feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.2, random_state=42)


feature_train = preprocessor.fit_transform(feature_train)


feature_test = preprocessor.transform(feature_test)


model = KNeighborsClassifier(n_neighbors=5)


model.fit(feature_train, target_train)


predictions = model.predict(feature_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

accuracy = accuracy_score(target_test, predictions)
precision = precision_score(target_test, predictions, average='weighted')
recall = recall_score(target_test, predictions, average='weighted')
f1 = f1_score(target_test, predictions, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


data = pd.read_csv('Drug.csv')


X = data[['Age',  'Na']]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


distortions = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    distortions.append(kmeans.inertia_)


plt.figure(figsize=(8, 4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Küme Sayısı')
plt.ylabel('İç Küme Hata Kareleri Toplamı (Inertia)')
plt.title('Dirsek Yöntemi')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


data = pd.read_csv('Drug.csv')


X = data[['Age', 'Na', ]] 


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)


cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_


print("Küme Merkezleri:")
print(cluster_centers)



# Görselleştirme
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', c='red', s=100)
plt.xlabel('Age')
plt.ylabel('Na')
plt.title('K-means Kümeleme')
plt.show()
```
